```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]
sales = [10, 20, 25, 30, 12, 15, 22]

plt.plot(days, sales,linestyle=':',marker='o',label='Sales',color='red')

plt.xlabel("Days")
plt.ylabel("Sales in each day")
plt.title("Basic Line Plot for Sales Analysis")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)

y1 = x**2
y2 = 2*x

plt.plot(x, y1, label='y = x²', linestyle=':', marker='o')
plt.plot(x, y2, label='y = 2x', linestyle='-', marker='s')

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Plot of y=x² and y=2x")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 21)
y = x * 2

plt.scatter(x, y, s=2, color='orange')
plt.title("Scatter Plot")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 16)
y = np.arange(10001,10016)
plt.scatter(x,y, s=50,facecolors='red',edgecolors='blue')
plt.title("Scatter plot")
plt.show()
```

```python
categories = ['A','B','C']
values = [10,20,15]
#plt.bar(categories,values)
plt.bar(categories,values,color='green',edgecolor='blue')
plt.title("Bar Chart")
plt.show()
```

```python
categories = ['A','B','C','D','E']
values = [10,20,15,20,35]
select=['green','red','purple','pink','black']
#plt.bar(categories,values,color=select,edgecolor='yellow',linewidth=5)
edge_select=['black','pink','red','purple','green']
plt.bar(categories,values,color=select,edgecolor=edge_select,linewidth=5)
plt.title("Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
weight = [30,35,42,60,67,45,39,24,73,49]
height = [132,130,142,149,162,182,142,155,148,164]
plt.scatter(weight,height, s=50,facecolors='yellow', edgecolors='green', label='Employees')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.grid(True)
plt.title("Height vs weight analysis")
plt.show()
```

```python
import matplotlib.pyplot as plt
products = ['apple','mango','orange','grapes','banana']
sales_amnt = [300,200,555,350,290]
select=['orange','blue','green','yellow','black']
plt.bar(products,sales_amnt,color=select, edgecolor='purple' , linewidth=4 , label='sales')
plt.xlabel("products")
plt.ylabel("sales_amnt")
plt.legend()
plt.title("Bar Chart")
plt.show()
```
