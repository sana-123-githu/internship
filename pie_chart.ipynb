```python
import matplotlib.pyplot as plt
sizes = [40,30,20,10]
labels = ['A','B','C','D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python

exp = [0,0,0.2,0]

plt.pie(sizes, labels=labels, autopct='%1.2f%%', explode=exp)
plt.title("Pie Chart")
plt.show()
```
