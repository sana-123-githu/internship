```python
import matplotlib.pyplot as plt
plt.subplot(2,2,1)
plt.plot([1,2,3],[4,5,6],marker='o')
plt.title("plot 1")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot([1,2,3],[6,5,4],marker='s')
plt.title("plot 2")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot([1,2,3],[2,3,1],marker='^')
plt.title("plot 3")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot([1,2,3],[3,1,2],marker='*')
plt.title("plot 4")
plt.grid(True)

plt.tight_layout()
plt.show()
```
